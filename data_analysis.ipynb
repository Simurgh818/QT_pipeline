```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
sns.set_theme()
```

```python
results_csv_path = r"C:\Users\sinad\OneDrive - Georgia Institute of Technology\CliffordandSameni\QT_results\QTdb_results\QT_wholedataset_results.csv"
df = pd.read_csv(results_csv_path)
df.head()
```

```python
df.describe()
```

```python
QTc1_median = df.loc[:,['channelNum', 'QTc1_median_human', 'QTc1_median_method_1','QTc1_median_method_2']]
QTc1_median
```

```python
QTc1_median_lead1 = QTc1_median[QTc1_median.loc[:, 'channelNum']==1]
QTc1_median_lead1
```

```python
QTc1_median_lead1 = QTc1_median_lead1.drop('channelNum', axis=1)
```

```python
sns.set(rc = {'figure.figsize':(8,4)})
ax = sns.boxplot(data=QTc1_median_lead1).set(title='Lead 1 Median QTlc Interval')
ax = sns.swarmplot(data=QTc1_median_lead1, color=".25")
# ax.fig.suptitle('Lead 1 Median QTlc Interval')
```

```python
QTc1_median_lead1.describe().round(decimals=4)
```

```python
sns.set(rc = {'figure.figsize':(8,4)})
ax2 = sns.boxplot(data=QTc1_median_lead2).set(title='Lead 2 Median QTlc Interval')
ax2 = sns.swarmplot(data=QTc1_median_lead2, color=".25")
```
