#### Method 2 Median RR is close to human in it's median value (50%), mean, standard deviation

```python
from sklearn.metrics import mean_squared_error
# QT.RMSE_method_1 = sqrt(nanmean((QT.QTc1_median_IQR_Method_1-QT.QTc1_median_human).^2));

RR_median_Method_1 = df.loc[:,['RR_median_method_1']]
RR_median_human =  df.loc[:,['RR_median_human']]
RR_median_Method_2 = df.loc[:,['RR_median_method_2']]
# RMSE_RR_median_method_1 = np.sqrt(np.nanmean(np.square(QTc1_median_IQR_Method_1 - QTc1_median_human)))
RMSE_RR_median_method_1 = mean_squared_error(RR_median_human, RR_median_Method_1, squared=False)
RMSE_RR_median_method_2 = mean_squared_error(RR_median_human, RR_median_Method_2, squared=False)
print('RMSE for Method 1 RR: ', np.round(RMSE_RR_median_method_1, 4), 'RMSE for Method 2 RR: ', np.round(RMSE_RR_median_method_2, 4))
# QTc1_median_human
```
